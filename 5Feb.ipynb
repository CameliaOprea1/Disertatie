```python
city_name = "Brasov, Romania"
G = ox.graph_from_place(city_name, network_type="drive", simplify=True)
```

```python
# Ensure the graph is unprojected
graph_unprojected = ox.project_graph(G, to_crs='epsg:4326')
# Add edge bearings-directions, to the unprojected graph
graph_with_bearings = ox.add_edge_bearings(graph_unprojected)
```

```python
for edge in graph_with_bearings.edges(data=True):
    nodes, data = edge[:2], edge[2]
    bearing_value = data['bearing']
    print(f'Edge {nodes}: Bearing Value - {bearing_value}')
```

```python
# Retrieve the boundary polygon for the city
city_boundary = ox.geocode_to_gdf(city_name)
boundary_polygon = city_boundary.geometry.unary_union

# Generate a random bounding box within the city boundary
minx, miny, maxx, maxy = boundary_polygon.bounds
random_minx = random.uniform(minx, maxx)
random_miny = random.uniform(miny, maxy)
random_maxx = random.uniform(random_minx, maxx)
random_maxy = random.uniform(random_miny, maxy)

random_bbox = (random_minx, random_miny, random_maxx, random_maxy)
print(random_bbox)
# Truncate the graph to only include nodes and edges within the random bounding box
G_random_part = ox.truncate.truncate_graph_bbox(G, north=random_maxy, south=random_miny, east=random_maxx, west=random_minx)
```

```python
# get n evenly-spaced colors from some matplotlib colormap
ec=ox.plot.get_colors(n=G_random_part.number_of_edges(), cmap="plasma", return_hex=True)
nc = ox.plot.get_node_colors_by_attr(G_random_part, attr="street_count", cmap="plasma")
fig, ax = ox.plot_graph(G_random_part, node_size=20,edge_color=ec, node_color=nc, bgcolor="grey",figsize=(14, 14))
```

```python
m = folium.Map(location=[random_miny + (random_maxy - random_miny) / 2, random_minx + (random_maxx - random_minx) / 2], zoom_start=14)

# Plot the graph on the folium map
ox.plot_graph_folium(G_random_part, graph_map=m, edge_color='grey', edge_width=2, edge_opacity=0.7)
for node, data in G_random_part.nodes(data=True):
    folium.CircleMarker([data['y'], data['x']], radius=5, color='blue', fill=True, fill_color='blue', fill_opacity=0.8, popup=str(node)).add_to(m)
m.save("truncatedGraph.html")
```

```python
df_coordinates = ox.graph_to_gdfs(graph_with_bearings, edges=False)["geometry"].apply(lambda x: (x.x, x.y)).reset_index()
df_coordinates.columns = ["Node_ID", "Coordinates"]
# Split the "Coordinates" column into separate "Latitude" and "Longitude" columns
df_coordinates[["Longitude","Latitude"]] = pd.DataFrame(df_coordinates["Coordinates"].tolist(), index=df_coordinates.index)
# Split the "Coordinates" column into separate "Latitude" and "Longitude" columns
df_coordinates[["Longitude","Latitude"]] = pd.DataFrame(df_coordinates["Coordinates"].tolist(), index=df_coordinates.index)

# Drop the now redundant "Coordinates" column
#df_coordinates = df_coordinates.drop(columns=["Coordinates"])
```

# Generating OD matrix

```python
n = 100
# Randomly select n origin and destination nodes
random_origin_nodes = np.random.choice(df_coordinates.index, min(n, len(df_coordinates)))
random_destination_nodes = np.random.choice(df_coordinates.index, min(n, len(df_coordinates)))
```

```python
randomized_df = pd.DataFrame({
    'Origin_Node_ID': random_origin_nodes,
    'Origin_Coordinates': df_coordinates.loc[random_origin_nodes, 'Coordinates'].values,
    'Destination_Node_ID': random_destination_nodes,
    'Destination_Coordinates': df_coordinates.loc[random_destination_nodes, 'Coordinates'].values
})
randomized_df[['Origin_Longitude', 'Origin_Latitude']] = pd.DataFrame(randomized_df['Origin_Coordinates'].tolist(), index=randomized_df.index)
randomized_df[['Destination_Longitude', 'Destination_Latitude']] = pd.DataFrame(randomized_df['Destination_Coordinates'].tolist(), index=randomized_df.index)
```

```python

mean_lat = randomized_df[['Origin_Latitude', 'Destination_Latitude']].values.flatten().mean()
mean_lon = randomized_df[['Origin_Longitude', 'Destination_Longitude']].values.flatten().mean()
mymap = folium.Map(location=[mean_lat, mean_lon], zoom_start=12)


# Create a MarkerCluster object
marker_cluster = MarkerCluster().add_to(mymap)

# Add markers and routes to the cluster
for index, row in enumerate(randomized_df.itertuples(), start=1):
    route_line = folium.PolyLine(
        locations=[[row.Origin_Latitude, row.Origin_Longitude],
                   [row.Destination_Latitude, row.Destination_Longitude]],
        color='gray',
        weight=2.5,
        opacity=0.8,
    )
    mymap.add_child(route_line)

    origin_marker = folium.Marker(
        location=[row.Origin_Latitude, row.Origin_Longitude],
        popup=f"Origin Nr crt: {index}",
        icon=folium.Icon(color='blue', prefix='fa', icon='circle')
    )
    origin_marker.add_to(marker_cluster)

    destination_marker = folium.Marker(
        location=[row.Destination_Latitude, row.Destination_Longitude],
        popup=f"Destination Nr crt: {index}",
        icon=folium.Icon(color='red', prefix='fa', icon='circle')
    )
    destination_marker.add_to(marker_cluster)

# Add a legend to the map
legend_html = """
     <div style="position: fixed; bottom: 50px; left: 50px; z-index:1000; font-size:14px; background-color:white; padding: 10px; border-radius: 5px; border: 2px solid gray;">
     &nbsp; <strong>Legend</strong> <br>
     &nbsp; <i class="fa fa-circle" style="color:blue"></i> Origin <br>
     &nbsp; <i class="fa fa-circle" style="color:red"></i> Destination
     </div>
     """
mymap.get_root().html.add_child(folium.Element(legend_html))
```
